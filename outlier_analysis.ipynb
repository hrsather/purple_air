## The random spikes that we see aren't all noise. 

```python
import matplotlib.pyplot as plt
import pandas as pd

import global_vars as gv
import load_data as ld
```

```python
def plot_ab(df_a, df_b):
    plt.plot(df_a[gv.DATE_NAME], df_a[gv.PPM_NAME])
    plt.plot(df_b[gv.DATE_NAME], df_b[gv.PPM_NAME])
    plt.legend(["A", "B"])
    plt.xlabel("Date and Time")
    plt.ylabel("ppm")
    plt.title(gv.PPM_NAME)
    plt.show()
```

```python
data = ld.get_raw_data("data")

arvada_a = data[data["location"] == "Arvada"]
arvada_b = data[data["location"] == "Arvada B"]
plot_ab(arvada_a, arvada_b)

# Zoom in a little closer
start_dt = "2021 08 11 22:00:00"
end_dt = "2021 08 13 23:00:00"
arvada_a_zoom = arvada_a[(arvada_a[gv.DATE_NAME] > pd.to_datetime(start_dt)) & (arvada_a[gv.DATE_NAME] < pd.to_datetime(end_dt))]
arvada_b_zoom = arvada_b[(arvada_b[gv.DATE_NAME] > pd.to_datetime(start_dt)) & (arvada_b[gv.DATE_NAME] < pd.to_datetime(end_dt))]
plot_ab(arvada_a_zoom, arvada_b_zoom)

# Close zoom in on real data
start_dt = "2021 08 12 21:50:00"
end_dt = "2021 08 12 22:40:00"
arvada_a_zoom = arvada_a[(arvada_a[gv.DATE_NAME] > pd.to_datetime(start_dt)) & (arvada_a[gv.DATE_NAME] < pd.to_datetime(end_dt))]
arvada_b_zoom = arvada_b[(arvada_b[gv.DATE_NAME] > pd.to_datetime(start_dt)) & (arvada_b[gv.DATE_NAME] < pd.to_datetime(end_dt))]
plot_ab(arvada_a_zoom, arvada_b_zoom)

print("Number of adjacent points over 100:", len(arvada_a_zoom[arvada_a_zoom[gv.PPM_NAME] > 100][gv.DATE_NAME]))
```

At a distance, the spikes appear to be outliers, both channels exhibit the spiking behaviour, so we can conclude that they are not, in fact, outliers, but true data readings.
